```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sample_data_01.csv")
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df[["総人口(男)"]]
y = df["総人口"]
model.fit(X, y)
```

```python
model.predict(np.array([[130000]]))
```

```python
model.coef_
```

```python
model.intercept_
```
